```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    res = np.cos(x)-2*x**3
    return res
def dfdx(x):
    res = -np.sin(x)-6*x**2
    return res
```

```python
max_iter=30
tol=0.000000000001
i=0
x0=-10
xi=x0

loop=[]
fonc=[]
x=[]

while abs(f(x0))>tol or i> max_iter:
    xi=x0-f(x0)/dfdx(x0)
    x0=xi
    i+=1
    loop.append(i)
    fonc.append(f(x0))
    x.append(x0)
    
plt.plot(x,fonc)
plt.plot(x[-1],fonc[-1], c='red',marker='o')
```

```python

x_plot = np.linspace(-2, 2, 100)
y_plot = f(x_plot)
```

```python
plt.plot(x_plot,y_plot)
plt.plot(xi, f(xi), c='red', marker='o', fillstyle='none')
```

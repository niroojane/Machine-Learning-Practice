```python

class Solution(object):
    def maxArea(self, height):
        
        vol =0
        left=0
        right=(len(height)-1)
        
        while left<right:
            if height[left]>=height[right]:
                vol=max(height[right]*(right-left),vol)
                right-=1
            else:
                vol=max(height[left]*(right-left),vol)
                left +=1
        return vol
```

```python
class Solution(object):
    def twosum(self, nums,target):
        dic=dict()
        for i in range(len(nums)):
            if nums[i] in dic:
                return [dic[nums[i]],i]
        
            else:
                dic[target - nums[i]] = i
                
    def twosum2(self,nums,target):
        nums=enumerate(nums)
        nums=sorted(nums, key=lambda x:x[1])
        
        l=0
        r=len(nums)-1
        
        while l<r:
            if nums[l][1]+nums[r][1]==target:
                return (nums[l][0],nums[r][0])
            elif nums[l][1]+nums[r][1]<target:
                l=l+1
            else:
                r=r-1
                
            
        
nums=[2,7,11,15]
target=9
Solution().twosum(nums,target)
```

```python
class Solution:
    def Roman(self,s):
        dic={'I': 1,'V': 5, 'X': 10,'L': 50,'C': 100,'D': 500, 'M': 1000}
        
        sum = 0
        for i in range(len(s)-1):
            if dic[s[i]]<dic[s[i+1]]:
                sum-=dic[s[i]]
            else:
                sum+=dic[s[i]]
        return sum +dic[s[-1]]
                
        
```

```python
strs=['flower','floor','fla','florere']

class Solution:
    def pref(self,strs):
        
        if not strs:
            return ''
        shortest=min(strs,key=len)
        for i,j in enumerate(shortest):
            for x in strs:
                if x[i]!=j:
                    return shortest[:i]
        return shortest
```

```python
class Solution:
    def searchRange(self, nums, target):
        def search(n):
            lo, hi = 0, len(nums)
            while lo < hi:
                mid = (lo + hi) // 2
                if nums[mid] >= n:
                    hi = mid
                else:
                    lo = mid + 1
            return lo
        lo = search(target)
        return [lo, search(target+1)-1] if target in nums[lo:lo+1] else [-1, -1]
```

```python
Solution().searchRange(nums,target)
```

```python
class Solution:
    def searchRange(self, nums, target):
        def binarySearchLeft(A, x):
            left, right = 0, len(A) - 1
            while left <= right:
                mid = (left + right) // 2
                if x > A[mid]: left = mid + 1
                else: right = mid - 1
            return left

        def binarySearchRight(A, x):
            left, right = 0, len(A) - 1
            while left <= right:
                mid = (left + right) // 2
                if x >= A[mid]: left = mid + 1
                else: right = mid - 1
            return right

        left, right = binarySearchLeft(nums, target), binarySearchRight(nums, target)
        return (left, right) if left <= right else [-1, -1]
```

```python
Solution().searchRange(nums,target)
```

```python
prices = [7,1,5,3,6,4]
class Solution:
    def maxprof(self,prices):
        maxprofit=0
        minprice=float('inf')
        for price in prices:
            minprice=min(minprice,price)
            profit=price-minprice
            maxprofit=max(maxprofit,profit)
        return maxprofit
 
    def pricer(self,prices):
        maxprofit=0
        left=0
        right=1
        
        while right<len(prices):
            profit=prices[right]-prices[left]
            if prices[left]<prices[right]:
                maxprofit=max(profit,maxprofit)
            else:
                left=right
            right=right+1
            
        return maxprofit
        
        
```

```python
class Solution:
    def threeSum(self, nums,):
  
        nums.sort()
        N, result = len(nums), []
        for i in range(N):
            if i > 0 and nums[i] == nums[i-1]:
                continue
            target = nums[i]*-1
            s,e = i+1, N-1
            while s<e:
                if nums[s]+nums[e] == target:
                    result.append([nums[i], nums[s], nums[e]])
                    s = s+1
                if nums[s] == nums[s-1]:
                    s = s+1
                elif nums[s] + nums[e] < target:
                    s = s+1
                else:
                    e = e-1
        return result
    
        
```

```python
class Solution:
    def threesum2(self,nums):
        res=set()
        n,p,z =[],[],[]
        for i in nums:
            if i==0:
                z.append(i)
            
            elif i<0:
                n.append(i)
            else:
                p.append(i)
                
        N=set(n)
        P=set(p)
        Z=set(z)
        
        if Z:
            for num in N:
                if -num in P:
                    res.add((-num,0,num))
        if len(Z)>3:
            res.add((0,0,0))
            
        for i in range(len(P)):
            for j in range(i+1,len(P)):
                target=(nums[i]+nums[j])*-1
                if target in N:
                    res.add(tuple(sorted([nums[i],nums[j],target])))
                    
        for i in range(len(N)):
            for j in range(i+1,len(N)):
                target=(nums[i]+nums[j])*-1
                if target in P:
                    res.add(tuple(sorted([nums[i],nums[j],target])))       
        return res
    
```

```python
class Solution:
    def maxprod(self,nums):
        
        ans=nums[0]
        maxprod=nums[0]
        minprod=nums[0]
        
        for i in range(1,len(nums)):
            x=max(nums[i],maxprod*nums[i],minprod*nums[i])
            y=min(nums[i],maxprod*nums[i],minprod*nums[i])
            maxprod=x
            minprod=y   
            ans=max(maxprod,ans)
        return ans
        
        
```

```python
class Solution:
    def parenth(self,strs):
        dic={"[":"]",'(':')','{':'}'}
        stack=[]
        
        for x in strs:
            if x in dic:
                stack.append(x)
            elif len(stack)==0 or dic[stack.pop()]!=x:
                return False
        return len(stack)==0
        
```

```python
class Solution:
    def rotate(self,matrix):
        r=len(matrix)-1
        l=0
        
        while l<r:
            matrix[l], matrix[r]= matrix[r], matrix[l]
            l=l+1
            r=r-1
        
        for i in range(len(matrix)):
            for j in range(i):
                matrix[i][j], matrix[j][i]= matrix[j][i], matrix[i][j]        
        
        
        
        return matrix
```
